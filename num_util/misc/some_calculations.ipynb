```python
import numpy as np
import astropy.units as u; import astropy.constants as c

m_HI = c.m_p+c.m_e
print('1 Msun / pc^3 corresponds to',   ( 1*u.Msun*u.pc**-3 / m_HI ).to( u.cm**-3 )  )
print('1 Msun / pc^3 = ', (1*u.Msun*u.pc**-3).to( u.g*u.cm**-3 ) )
print('1 pc / Myr=',  (1*u.pc/u.Myr).to(u.km/u.s)  )
```
